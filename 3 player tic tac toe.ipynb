```python
# Initialize the game board
game_board = [[' ' for _ in range(6)] for _ in range(6)]

# Function to display the game board
def display_board(board):
    """Displays the current state of the game board."""
    for row in board:
        print('|'.join(row))
    print('-' * 11)

# Function to check if a player has won
def check_victory(board, player):
    """Checks if the given player has won."""
    # Check rows and columns for victory
    for i in range(6):
        if ''.join(board[i]).count(player) >= 4 or ''.join([row[i] for row in board]).count(player) >= 4:
            return True
    # Check diagonals for victory
    for i in range(3):
        if ''.join([board[j][i+j] for j in range(6-i)]).count(player) >= 4 or ''.join([board[i+j][j] for j in range(6-i)]).count(player) >= 4:
            return True
    return False

# Function to simulate the game
def play_game():
    """Simulates a game."""
    players = ['X', 'O', 'T']  # Players: X, O, T
    current_player_index = 0  # Start with player X
    while True:
        current_player = players[current_player_index]
        display_board(game_board)
        print("Player", current_player + "'s turn")
        row = int(input("Enter the row (1-6): ")) - 1
        col = int(input("Enter the column (1-6): ")) - 1
        # Check if the selected cell is empty
        if game_board[row][col] == ' ':
            game_board[row][col] = current_player
            if check_victory(game_board, current_player):
                display_board(game_board)
                print("Player", current_player, "wins!")
                break
            # Check for a draw
            elif all(cell != ' ' for row in game_board for cell in row):
                display_board(game_board)
                print("It's a draw!")
                break
            # Switch to the next player
            current_player_index = (current_player_index + 1) % 3
        else:
            print("That cell is already occupied. Please choose another.")

# Start the game
play_game()
```
